## Key Idea

A domain model keeps core business logic independent from infrastructure like databases and APIs.

## Core Concepts

### 1. Domain Model
- **Definition:** A structured representation of business logic using classes and objects.
- **Problem it solves:** Prevents business logic from being scattered across the codebase.
- **When to use:** When building applications with complex rules or workflows.
- **When not to use:** For very small scripts or simple one-file programs
