## What is a Domain Model?

> "The domain is a fancy way of saying *the problem you’re trying to solve*."

- The **Domain Model** is our code representation of the business rules and mental model used by domain experts.
- It should be **independent** of databases, web frameworks, and external tools.
- Goal: Make business logic highly testable and maintainable.

## Key Patterns in Domain Modeling

### 1. Value Object
- Identified by its **data** (not by identity)
- Usually **immutable**
- Example: Money, DateRange, OrderReference

### 2. Entity
- Has a **long-lived identity** (e.g., ID)
- Mutable over time
- Example: Batch, Order, User

### 3. Domain Service
- Business logic that doesn’t naturally fit inside an Entity or Value Object
- Often stateless functions or small classes

```python
# Example from the book - Value Object
from dataclasses import dataclass
from datetime import date

@dataclass(frozen=True)   # immutable
class OrderLine:
    orderid: str
    sku: str
    qty: int

# Usage
line1 = OrderLine("order-001", "SKU-123", 10)
line2 = OrderLine("order-001", "SKU-123", 10)

print(line1 == line2)   # True (because data is the same)
```

```python
# Example - Entity (Batch)
from dataclasses import dataclass
from datetime import date
from typing import Optional

@dataclass
class Batch:
    reference: str
    sku: str
    qty: int
    eta: Optional[date] = None
    
    def allocate(self, line: OrderLine) -> None:
        """Allocate stock to an order line"""
        if self.can_allocate(line):
            self.qty -= line.qty

    def can_allocate(self, line: OrderLine) -> bool:
        return self.sku == line.sku and self.qty >= line.qty
```

## Key Learnings from Chapter 1

- Domain Model should focus purely on **business rules**
- Use **Value Objects** for things defined by their attributes
- Use **Entities** when identity matters more than current values
- Start simple and make it testable first
- This layer is the **heart** of Clean Architecture
