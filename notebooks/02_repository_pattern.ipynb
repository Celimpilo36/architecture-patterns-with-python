**Goal**: Decouple our Domain Model from the database implementation.

## Why Do We Need the Repository Pattern?

- Domain layer should **not** know about SQLAlchemy, databases, or queries.
- Repository acts as a **collection interface** for our domain objects.
- It hides persistence details behind a clean API.
- Makes testing much easier (we can use FakeRepository).

## Core Ideas

- **AbstractRepository** → Defines the interface (what the domain can do)
- **SqlAlchemyRepository** → Real implementation (talks to database)
- **FakeRepository** → Used for unit tests

```python
# src/domain/repository.py
from abc import ABC, abstractmethod
from typing import List, Optional
from src.domain.model import Batch

class AbstractRepository(ABC):

    @abstractmethod
    def add(self, batch: Batch) -> None:
        pass

    @abstractmethod
    def get(self, reference: str) -> Optional[Batch]:
        pass

    @abstractmethod
    def list(self) -> List[Batch]:
        pass
```

```python
# src/adapters/repository.py
from src.domain.repository import AbstractRepository
from src.domain.model import Batch
from sqlalchemy.orm import Session

class SqlAlchemyRepository(AbstractRepository):

    def __init__(self, session: Session):
        self.session = session

    def add(self, batch: Batch):
        self.session.add(batch)

    def get(self, reference: str) -> Optional[Batch]:
        return self.session.query(Batch).filter_by(reference=reference).one_or_none()

    def list(self) -> List[Batch]:
        return self.session.query(Batch).all()
```

## Key Learnings - Chapter 2

- Repository Pattern hides data access logic
- Domain depends on **AbstractRepository** (Dependency Rule)
- Concrete repositories (SQLAlchemy) live in `adapters/`
- Fake repositories make unit tests fast and reliable
- This is a major step toward **decoupling**
