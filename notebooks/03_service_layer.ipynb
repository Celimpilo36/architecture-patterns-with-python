## What is the Service Layer?

- Acts as the **orchestrator** between the Entrypoint and the Domain.
- Contains **use cases** / application logic.
- Keeps domain model pure and entrypoints thin.
- Depends on **AbstractRepository** (following Dependency Rule).

## Key Responsibilities

- Fetch data using Repository
- Coordinate domain objects
- Handle business logic that spans multiple domain objects
- Raise domain exceptions

````python
# src/service_layer/services.py
from src.domain.model import OrderLine
from src.domain.repository import AbstractRepository

class InvalidSku(Exception):
    """Raised when SKU is not found"""

def allocate(line: OrderLine, repo: AbstractRepository) -> str:
    """Allocate stock to an order line"""
    
    batches = repo.list()
    
    try:
        batch = next(b for b in batches if b.can_allocate(line))
        batch.allocate(line)
        return batch.reference
    except StopIteration:
        raise InvalidSku(f"Invalid sku {line.sku}")
````

## Key Learnings - Chapter 3

- Service Layer contains **use cases**
- It depends on `AbstractRepository` (Dependency Inversion)
- Makes business logic easy to test with `FakeRepository`
- Raises clear domain exceptions (`InvalidSku`)
- Prepares us for Unit of Work pattern (next chapters)
